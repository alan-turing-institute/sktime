# Exploratory data analysis with sktime

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sktime.utils.plotting import plot_lags
```

## Plotting lagged versions of a time series
- Lag plots of a time series are scatter plots of corresponding lagged versions of a time series. Trends in lag plots give use insight into the autocorrelated behaviour i.e. patterns in time series. Presence of such patterns are a useful indication for predictability of the time series as well as for identifying outliers from the regular pattern.
- In the example below we plot lagged versions of a random walk with the `plot_lags` function.     


```python
# Creating a sample time series for testing
np.random.seed(42)
x = np.cumsum(np.random.normal(loc=1, scale=5, size=50))
s = pd.Series(x)
```

```python
fig = plt.figure(figsize=(8, 6 * 1))
ax = fig.add_subplot(1, 1, 1)
plt.plot(s)
```

```python
# Case1: Plotting a single lag of a time series
fig, axes = plot_lags(s, 2)
```

```python
# Case2: Plotting several time series lags
fig, axes = plot_lags(s, [1, 2, 3, 4, 5])
```
